##Ejemplo 3. Histograma de 2 grupos

1. En este ejemplo generamos 100 numeros aleatorios reales que representa la producción diaria de 100 telares (medidas en metros).
2. Generar otera lista con la producción de 100 telares para una segunda fabrica.
3. Con esta lista de valores se pretende determinar las medidas de tendencia central, la media, la moda, la mdeiana, la desviación estandar.
4. Graficar un histograma para cada fabrica.

```python
#Libreria para generar valores aleatorios
import random
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio2 = round(random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
#Libreria de pyton para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python
media=mean(fabrica1)
mediana=median(fabrica1)
moda=mode(fabrica1)
print("Valores de la fabrica 1")
print("La media=", media)
print("La mediana es =", mediana)
print("La moda es =", moda)
```

```python
media=mean(fabrica2)
mediana=median(fabrica2)
moda=mode(fabrica2)
print("Valores de la fabrica 2")
print("La media=", media)
print("La mediana es =", mediana)
print("La moda es =", moda)
```

##Tablas de frecuencia

```python
# generamos una lista vacia que repsresenta un grupo de personas
edades=[]
for i in range(100):
  numero_aleaotrio=random.randint(18,70)
  edades.append(numero_aleaotrio)
print (edades)
```

```python
import pandas as pd
df_edades = pd.DataFrame(edades, columns=["Edad"])
df_edades
```

```python
#Calcular la tabla de distribucion de frecuencias
tabla_frecuencias=df_edades["Edad"].value_counts().reset_index()
tabla_frecuencias.columns=["Edad", "Frecuencia"]
tabla_frecuencias=tabla_frecuencias.sort_values(by="Edad")
tabla_frecuencias
```

```python
tabla_frecuencias["Frec. Relativa"]= tabla_frecuencias["Frecuencia"]/100
```

```python
tabla_frecuencias
```

```python
#Calcular la frecuenica porcental
tabla_frecuencias["Frec. porcentual"]=tabla_frecuencias["Frecuencia"]/100*100
```

```python
tabla_frecuencias
```

```python
#Calcular la frecuencia acumulada
#Agregamos una nueva columna con el calculo
tabla_frecuencias["Frec. acumulada"]=tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```

##Regla de sturges

```python
#Libreria de funciones matematias
import math
```

```python
print(edades)
```

```python
plt.hist(edades, alpha=0.5,rwidth=.95)
plt.title("Histograma de edades basado en valores estandar de matplotlib")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
```

```python
df_edades_sturges= pd.DataFrame({"Edad":edades})
df_edades_sturges
```

```python
#Calcular el numero de interlvalos o clases utilizando la regla de sturges
n=len(edades)
k=int(1+math.log2(n))
print("Valor de n= ",n,"\nValor de k o numero de intervalso o clases= ",k)
```

```python
tabla_frec_sturges=pd.cut(df_edades_sturges["Edad"],bins=k).value_counts().reset_index()
```

```python
tabla_frec_sturges
```

```python
tabla_frec_sturges.columns=["Clase","Frecuencia"]
tabla_frec_sturges
```

```python
#Ordenar la tabla de frecuencia por la clase o el intervalo
tabla_frec_sturges=tabla_frec_sturges.sort_values(by="Clase")
tabla_frec_sturges
```

```python
tabla_frec_sturges["Frec. relativa"]=tabla_frec_sturges["Frecuencia"]/n
tabla_frec_sturges["Frec. porcentual"]=tabla_frec_sturges["Frecuencia"]/n*100
tabla_frec_sturges["Frec. acumulada"]=tabla_frec_sturges["Frecuencia"].cumsum()
```

```python
tabla_frec_sturges
```

```python
plt.hist(df_edades_sturges["Edad"],bins=k,alpha=.5,rwidth=.95)
plt.title("Histograma de edades basado en la regla de sturges")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
```
